```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
n=1000

data = pd.DataFrame({
    'timestamp': pd.date_range(start='2022-01-01', periods=n, freq='H'),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n),
    'latency_ms' : np.random.normal(50,10,n),   # (loc=mean,scale=sd,size=n)
    'packet_loss_%' : np.random.uniform(0,5,n), #(low,high,size)
    'throughput_mbps': np.random.normal(100,20,n),
    'network_type': np.random.choice(['4G', '5G'],n)}
)
```

```python
data.to_csv('network_data.csv', index=False)
```
